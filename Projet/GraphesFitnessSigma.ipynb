```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
data1 = pd.read_csv("Data/DiscretTrous/DiscreteFlipTrous1.txt", sep=";")
data2 = pd.read_csv("Data/DiscretTrous/DiscreteFlipTrous2.txt", sep=";")

data1.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,6))

Values = [(0, 0, -3000, "k")]
mean = Values[0][0]
stdev = Values[0][1]
treshold = Values[0][2]

dataf1  = data1[((data1["E(fitness)"] >= treshold) & (data1["mean"] == mean) & (data1["stdev"] == stdev))]
axs[0].scatter(dataf1["Generation"], dataf1["E(fitness)"], c="k", alpha=0.5, s=10, label="Srie 1")

dataf2  = data2[((data2["E(fitness)"] >= treshold) & (data2["mean"] == mean) & (data2["stdev"] == stdev))]
axs[0].scatter(dataf2["Generation"], dataf2["E(fitness)"], c="tab:blue", alpha=0.5, s=10, label="Srie 2")

axs[0].set(xlabel="Generation", ylabel="E(fitness)", title="E(fitness)")

dataf1  = data1[((data1["sigma(fitness)"] < 6000) & (data1["mean"] == mean) & (data1["stdev"] == stdev))]
axs[1].scatter(dataf1["Generation"], dataf1["sigma(fitness)"], c="k", alpha=0.5, s=10, label="Srie 1")

dataf2  = data2[((data2["sigma(fitness)"] < 6000) & (data2["mean"] == mean) & (data2["stdev"] == stdev))]
axs[1].scatter(dataf2["Generation"], dataf2["sigma(fitness)"], c="tab:blue", alpha=0.5, s=10, label="Srie 2")

axs[1].set(xlabel="Generation", ylabel="sigma(fitness)", title="sigma(fitness)")
axs[1].legend()


plt.show()
plt.show()
```

```python
data1 = pd.read_csv("Data/Gaz/TestGaz.txt", sep=";")
data2 = pd.read_csv("Data/Gaz/TestGazExp.txt", sep=";")


data1.head()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12,6))

axs[0].plot(data1["Generation"], data1["E(fitness)"], c="k", label="data")
axs[0].fill_between(data1["Generation"], data1["E(fitness)"] - data1["sigma(fitness)"], data1["E(fitness)"] + data1["sigma(fitness)"], alpha=0.5, color="gray", label="Incertitudes")
axs[0].set(xlabel="Generation", ylabel="E(fitness)", title="TestGaz")
axs[0].plot(data1["Generation"], data1["Generation"]-data1["Generation"], linestyle='--', color = "tab:red", label="Gaz Ideal ($n_s=1$)")

axs[1].plot(data2["Generation"], data2["E(fitness)"], c="k", label="data")
axs[1].fill_between(data2["Generation"], data2["E(fitness)"] - data2["sigma(fitness)"], data2["E(fitness)"] + data2["sigma(fitness)"], alpha=0.5, color="gray", label="Incertitudes")
axs[1].set(xlabel="Generation", ylabel="E(fitness)", title="TestGazExp")
axs[1].plot(data2["Generation"], data2["Generation"]-data2["Generation"] +10, linestyle='--', color = "tab:red", label="Gaz Ideal ($n_s=1$)")

axs[0].legend()
axs[1].legend()
plt.show()
```

```python
def Visualiser_data(mean, stdev, acceptation, Name, data):
    A = np.array([])
    B = [str(mean), str(stdev), str(acceptation)]

    df = data[((data["mean"] == mean) & (data["stdev"] == stdev) & (data["acceptation"] == acceptation))]

    mask = (df["Generation"] == 0)

    A = np.array([df[mask]["E(fitness)"], df[mask]["fitness1"], df[mask]["Genes(Scores(0,1),:)"]])
    B.append(str(A[0][0]))
    B.append(str(A[1][0]))
    B.append(A[2,0][2:-2].replace(",",""))


    mask = (df["fitness1"] == df["fitness1"].max())

    A = np.array([df[mask]["E(fitness)"], df[mask]["fitness1"], df[mask]["Genes(Scores(0,1),:)"]])
    B.append(str(A[0][0]))
    B.append(str(A[1][0]))
    B.append(A[2,0][2:-2].replace(",",""))

    mask = (df["Generation"] == df["Generation"].max())

    A = np.array([df[mask]["E(fitness)"], df[mask]["fitness1"], df[mask]["Genes(Scores(0,1),:)"]])
    B.append(str(A[0][0]))
    B.append(str(A[1][0]))
    B.append(A[2,0][2:-2].replace(",",""))

    with open(Name, "w") as file:
        # Write each string to the file
        for string in B:
            file.write(string + "\n")
```

```python
#stdev = 10 -> mean = -10, 0, 10
#stdev = 8 -> -5,0, 5

mean = 0
stdev = 0
acceptation = 0.9
Name = "test.txt"
Visualiser_data(mean, stdev, acceptation, Name, data1)
```
